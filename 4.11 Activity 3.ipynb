```python
import pandas as pd
import string 
import random
import numpy as np
df = pd.read_csv("PeoplesFavourites-1 copy.csv")
```

```python
df.replace('none', np.nan, inplace=True)
df.replace("0", np.nan, inplace=True)
df.replace(" ", np.nan, inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df.dropna(subset=["First Name","Last Name","Email"], inplace = True)
```

```python
s = list()
```

```python
number = df.index 
letters = list(string.ascii_uppercase)
numbers = list(string.digits)
```

```python
for i in range(df.count().max()):
    number_uniq = number[i] + 10000
    letter = random.choice(letters)
    number2 = random.choice(numbers)
    uniq_id = str(number_uniq)[1:] + letter + number2
    s.append(uniq_id)
```

```python
pd.Series(s).nunique() == df.count().max()
```

```python
df["First Name"] = s
df.rename(columns = {'First Name':'Unique ID'}, inplace = True) 
df.drop(columns="Last Name",inplace=True)
```

```python
df.set_index("Unique ID",inplace=True)
```

```python
df.sort_values('Income', ascending=False, inplace=True)
```

```python
df.dropna(subset=["Interests","Interests.1","Food","Food.1","Car","Movie","Movie.1"], how="all", inplace=True)
```
